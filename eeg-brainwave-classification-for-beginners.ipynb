# Split the data

```python
target = data['label']
data1 = data.drop(['label'], axis=1)
```

<span style="font-family:Verdana;"> <b>  we seperate the target column from the data </b></span>

# Scale the data

```python
scaler = StandardScaler()
data2 = pd.DataFrame(scaler.fit_transform(data1))
```

<span style="font-family:Verdana;"> <b> The standard score of a sample x is calculated as:

z = (x - u) / s
    </b></span>

```python
data2
```

# Encode the target

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
y1 = pd.DataFrame(enc.fit_transform(target))
```

<span style="font-family:Verdana;text-align:center;"> <b>We encode the target
  <ul>
      <li><span  >Neutral   --> 0 </span></li>
       <li><span >Negative --> 1 </span></li>
       <li><span  >Positive --> 2 </span></li>
    </ul>
    </b></span>

```python
y1
```

<span style="font-family:Verdana;"> <b> We split the data into test and train.
    <br>
80% for training and 20% for the test .</b></span>

```python
X_train,X_test, y_train, y_test = train_test_split(data2, y1, train_size = 0.8)
```

# Modeling

```python
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(128, 128), random_state=1)
clf.fit(X_train, y_train)
```

# Accuracy of the model

```python
accuracy = clf.score(X_test, y_test)
print(f'Accuracy: {accuracy}')
```

<span style="font-family:Verdana;"> <b> Visualize the confusion matrix </b></span>

```python
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import ConfusionMatrixDisplay
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_test))
disp.plot(cmap=plt.cm.Blues)
disp.ax_.set_title('Confusion Matrix')

plt.show()
```

```python
clr = classification_report(y_test, y_pred, target_names={"Neutral","Positive","Negative"})
print("Classification Report:\n----------------------\n", clr)
```
